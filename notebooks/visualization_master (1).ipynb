```python
#Refferred back to earlier labs/projects for visualization techniques
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read in csv file
df = pd.read_csv('Datasets/train.csv')
```

```python
#Heatmap for all the continous variables
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot = True)
```

```python
#Plotting distribution of budget
sns.displot(df['budget'])
plt.title('Budget Distribution')
```

```python
#Plotting distribution of popularity
sns.displot(df['popularity'])
plt.title('Popularity Counts')
```

```python
#Building scatter plot to see relationship between popularity and revenue
plt.scatter(df['popularity'], df['revenue'])
plt.xlabel('Popularity')
plt.ylabel('Revenue')
plt.title('Scatter plot of Revenue vs Popularity')
```

```python
#Building scatter plot to see relationship between budget and revenue
plt.title('Plot of Budget against Revenue')
sns.regplot(data=df, x='budget', y='revenue')
```

```python
#Histogram for movie length
plt.hist(df['runtime'], bins = 25)
plt.title('Histogram for length of Movies')
plt.xlabel('Length in minutes')
plt.ylabel('Number of occurences')
```

```python
#Bar chart for number of movies released per year

#Take last 2 numbers from date which corresponds to year
df['year'] = df['release_date'].str[-2:].astype(int)

#Add 2000 to years between 00-17, and 1900 to anything else
df['year'] = df['year'].apply(lambda x: 1900 + x if x > 17 else 2000 + x)

# Count the occurrences of each unique year
#Using sort_index to make sure it follows in order of year rather than size of occurences
year_counts = df['year'].value_counts().sort_index()

# Create a bar graph
plt.figure(figsize=(10, 10))
year_counts.plot(kind='bar')
plt.title('Number of Movie Releases per Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.xticks(rotation=90)
plt.show()
```
